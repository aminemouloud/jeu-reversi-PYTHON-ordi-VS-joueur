```python
#version 1 partie avec dico
# a conserver pour jouer contre l ordi

def newLigne(nbColonnes, valeurParDefaut=0):
    liste=[]
    for colonne in range(0,nbColonnes):
        liste.append(valeurParDefaut)
    return(liste)

def newMatrice(nbLignes, nbColonnes, valeurParDefaut=0):
    matrice=[]
    for ligne in range(0,nbLignes):
        ma_ligne=newLigne(nbColonnes, valeurParDefaut)
        matrice.append(ma_ligne)
    return(matrice)

def afficherPremiereLigne(nbColonnes,tailleCellule=4):
    s=''.center(tailleCellule)+'|'
    for colonne in range(0, nbColonnes):
        s=s+str(colonne).center(tailleCellule)+'|'
    print(s)
    
def getNbLignes(matrice):
    return(len(matrice))

def getNbColonnes(matrice):
    return(len(matrice[0]))

def setVal(matrice, ligne, colonne, valeur):
    matrice[ligne][colonne]=valeur
    
def getVal(matrice, ligne, colonne):
    return(matrice[ligne][colonne])

def afficherLigne(numero_ligne, ligne, tailleCellule=4):
    s=str(numero_ligne).center(tailleCellule)+'|'
    for colonne in range(0, len(ligne)):
        s=s+str(ligne[colonne]).center(tailleCellule)+'|'
    print(s)

def interligne(mat, tailleCellule):
    s='----+'.center(tailleCellule)
    for colonne in range(0, len(mat[0])):
        s=s+'----+'.center(tailleCellule)
    print(s)    
    
def afficherMatrice(mat, tailleCellule=4):
    nbColonnes=getNbColonnes(mat)
    afficherPremiereLigne(nbColonnes,tailleCellule) 
    interligne(mat, tailleCellule)
    for ligne in range(0,len(mat)):
        numero_ligne=ligne
        ma_ligne=mat[ligne]
        afficherLigne(numero_ligne, ma_ligne, tailleCellule)
        interligne(mat, tailleCellule)

def capture(joueur):
    if (joueur==1):
        return(2)
    else:
        return(1)

def occurences(plateau, nombre):
    nbLignes=getNbLignes(plateau)
    nbColonnes=getNbColonnes(plateau)
    s=0
    for ligne in range(0,nbLignes):
        for colonne in range(0,nbColonnes):
            if (plateau[ligne][colonne]==nombre):
                s=s+1
    return s
    
def qui_a_gagné(plateau, dico):
    if (occurences(plateau,2)<occurences(plateau,1)):
        print('Le gagnant est  joueur 1 ')
    elif (occurences(plateau,1)<occurences(plateau,2)):
        print('Le gagnant est  joueur 2  ')
    else:
        print('Aucun gagnant !')
        
        
def traiter_horizontal(plateau,ligne,colonne, joueur, 
                       liste_ok, ordi, listeLongueur):

    nbColonnes=getNbColonnes(plateau)
    val_capture=capture(joueur)
    
    
    #sens colonne croissante
    col=colonne+1
    ok=False
    liste=[]
    while (col!=nbColonnes) and not(ok):
        if (plateau[ligne][col]==val_capture):
            liste.append(col)
            col=col+1
        elif (plateau[ligne][col]==joueur):
            ok=True
        else:
            col=nbColonnes
            
    if not(ok):
        liste=[]

    if (len(liste)!=0):
        liste_ok[0]=True
    
    if not(ordi):
        for col in liste:
            plateau[ligne][col]=joueur
    
    listeLongueur[0]=listeLongueur[0]+len(liste)
        
    #sens colonne décroissante
    col=colonne-1
    ok=False
    liste=[]
    while (col!=-1) and not(ok):
        if (plateau[ligne][col]==val_capture):
            liste.append(col)
            col=col-1
        elif (plateau[ligne][col]==joueur):
            ok=True
        else:
            col=-1
            
    if not(ok):
        liste=[]

    if (len(liste)!=0):
        liste_ok[0]=True
    
    if not(ordi):
        for col in liste:
            plateau[ligne][col]=joueur
            
    listeLongueur[0]=listeLongueur[0]+len(liste)
        

def traiter_vertical(plateau,ligne,colonne, joueur, liste_ok, ordi, listeLongueur):

    nbLignes=getNbLignes(plateau)
    nbColonnes=getNbColonnes(plateau)
    val_capture=capture(joueur)
    
    #sens ligne croissante
    lig=ligne+1
    ok=False
    liste=[]
    while (lig!=nbLignes) and not(ok):
        if (plateau[lig][colonne]==val_capture):
            liste.append(lig)
            lig=lig+1
        elif (plateau[lig][colonne]==joueur):
            ok=True
        else:
            lig=nbLignes
            
    if not(ok):
        liste=[]
    
    if (len(liste)!=0):
        liste_ok[1]=True
    
    if not(ordi):
        for lig in liste:
            plateau[lig][colonne]=joueur
    
    listeLongueur[0]=listeLongueur[0]+len(liste)
    
    #sens décroissante
    lig=ligne-1
    ok=False
    liste=[]
    while (lig!=-1) and not(ok):
        if (plateau[lig][colonne]==val_capture):
            liste.append(lig)
            lig=lig-1
        elif (plateau[lig][colonne]==joueur):
            ok=True
        else:
            lig=-1
            
    if not(ok):
        liste=[]
        
    if (len(liste)!=0):
        liste_ok[1]=True
    
    if not(ordi):
        for lig in liste:
            plateau[lig][colonne]=joueur
    
    listeLongueur[0]=listeLongueur[0]+len(liste)
        
def traiter_diagonale_PPLE(plateau,ligne,colonne, joueur, liste_ok, ordi, listeLongueur):
    
    nbLignes=getNbLignes(plateau)
    nbColonnes=getNbColonnes(plateau)
    val_capture=capture(joueur)
    
    #sens ligne croissante
    lig=ligne+1
    col=colonne+1
    ok=False
    liste=[]
    while (lig!=nbLignes) and col!=nbColonnes and not(ok):
        if (plateau[lig][col]==val_capture):
            liste.append((lig,col))
            lig=lig+1
            col=col+1
        elif (plateau[lig][col]==joueur):
            ok=True
        else:
            lig=nbLignes
            
    if not(ok):
        liste=[]
        
    if (len(liste)!=0):
        liste_ok[2]=True           

    if not(ordi):
        for e in liste:
            lig=e[0]
            col=e[1]
            plateau[lig][col]=joueur
            
    listeLongueur[0]=listeLongueur[0]+len(liste)   
    #sens ligne colonne décroissantes
    lig=ligne-1
    col=colonne-1
    ok=False
    liste=[]
    while (lig!=-1) and (col!=-1) and not(ok):
        if (plateau[lig][col]==val_capture):
            liste.append((lig,col))
            lig=lig-1
            col=col-1
        elif (plateau[lig][col]==joueur):
            ok=True
        else:
            lig=-1
            
    if not(ok):
        liste=[]
        
    if (len(liste)!=0):
        liste_ok[2]=True
    
    if not(ordi):
        for e in liste:
            lig=e[0]
            col=e[1]
            plateau[lig][col]=joueur
            
    listeLongueur[0]=listeLongueur[0]+len(liste)
        
def traiter_diagonale_SEC(plateau,ligne,colonne, joueur, liste_ok, ordi, listeLongueur):
    
    nbLignes=getNbLignes(plateau)
    nbColonnes=getNbColonnes(plateau)
    val_capture=capture(joueur)
    
    #sens descendant
    lig=ligne+1
    col=colonne-1
    ok=False
    liste=[]
    while (lig!=nbLignes) and col!=-1 and not(ok):
        if (plateau[lig][col]==val_capture):
            liste.append((lig,col))
            lig=lig+1
            col=col-1
        elif (plateau[lig][col]==joueur):
            ok=True
        else:
            lig=nbLignes
            
    if not(ok):
        liste=[]
        
    if (len(liste)!=0):
        liste_ok[3]=True
   
    if not(ordi):
        for e in liste:
            lig=e[0]
            col=e[1]
            plateau[lig][col]=joueur
    
    listeLongueur[0]=listeLongueur[0]+len(liste)
    #sens montant
    lig=ligne-1
    col=colonne+1
    ok=False
    liste=[]
    while (lig!=-1) and (col!=nbColonnes) and not(ok):
        if (plateau[lig][col]==val_capture):
            liste.append((lig,col))
            lig=lig-1
            col=col+1
        elif (plateau[lig][col]==joueur):
            ok=True
        else:
            lig=-1
            
    if not(ok):
        liste=[]
    
    if (len(liste)!=0):
        liste_ok[3]=True
    
    if not(ordi):
        for e in liste:
            lig=e[0]
            col=e[1]
            plateau[lig][col]=joueur
            
    listeLongueur[0]=listeLongueur[0]+len(liste)

# traiter modifie le tableau quand joueur arrive à la case
# plateau[ligne][colonne]
def traiter(plateau, ligne, colonne, joueur,liste_ok, ordi, listeLongueur):
    if (plateau[ligne][colonne]==0):
        traiter_horizontal(plateau, ligne, colonne, joueur, liste_ok, ordi, listeLongueur)
        traiter_vertical(plateau, ligne, colonne, joueur, liste_ok, ordi, listeLongueur)
        traiter_diagonale_PPLE(plateau,ligne,colonne, joueur, liste_ok, ordi, listeLongueur)
        traiter_diagonale_SEC(plateau,ligne,colonne, joueur, liste_ok, 
                              ordi, listeLongueur)
    else:
        liste_ok=[False,False,False,False]
    
    
def partie(plateau, tailleCellule):

    print('C est joueur 1 qui commence à jouer')
    joueur=1
    ordi=True
    ok1=False
    ok2=False
    ok=True
    nbre_tours=0
    listeLongueur=[0]
    afficherMatrice(plateau,tailleCellule)
    while (ok):
        nbre_tours=nbre_tours+1
        liste_ok=[False, False, False, False]
        if (joueur==1) and (ordi==True):
            maxiC=0
            li_max=0
            col_max=0
            for ligne in range(0, nbLignes):
                for colonne in range(0, nbColonnes):
                    listeLongueur=[0]
                    liste_ok=[False, False, False, False]
                    if (plateau[ligne][colonne]==0):
                        traiter(plateau, ligne, colonne, joueur, liste_ok, 
                                ordi, listeLongueur)
                        if (maxiC<listeLongueur[0]):
                            maxiC=listeLongueur[0]
                            li_max=ligne
                            col_max=colonne
            if (0<maxiC):             
                ordi=False
                joueur=1
                listeLongueur=[0]
                liste_ok=[False, False, False, False] 
                traiter(plateau, li_max, col_max, joueur, liste_ok, ordi, listeLongueur)
                setVal(plateau, li_max, col_max, joueur)  
                joueur=2
                ordi=True
            else: 
                ok=False
                ok1=True
                
        elif (joueur==2) and (ordi==True):
            maxiC=0
            li_max=0
            col_max=0
            for ligne in range(0, nbLignes):
                for colonne in range(0, nbColonnes):
                    listeLongueur=[0]
                    liste_ok=[False, False, False, False]
                    if (plateau[ligne][colonne]==0):
                        traiter(plateau, ligne, colonne, joueur, liste_ok, 
                                ordi, listeLongueur)
                        if (maxiC<listeLongueur[0]):
                            maxiC=listeLongueur[0]
                            li_max=ligne
                            col_max=colonne
            if (0<maxiC):               
                ordi=False
                joueur=2
                listeLongueur=[0]
                liste_ok=[False, False, False, False] 
                traiter(plateau, li_max, col_max, joueur, liste_ok, ordi, listeLongueur)
                setVal(plateau, li_max, col_max, joueur)  
                joueur=1
                ordi=True
            else:
                ok=False
                ok2=True
    
    print(' Apres ', nbre_tours, ' tours ')
    if (ok1):
        print(' Le gagnant est joueur 2 car joueur 1 n pas pu jouer ')
        print(' Le nombre de pions de joueur 1 est ', occurences(plateau,1))
        print(' Le nombre de pions de joueur 2 est ', occurences(plateau,2))
    elif (ok2):
        print(' Le gagnant est joueur 1 car joueur 2 n pas pu jouer ')
        print(' Le nombre de pions de joueur 1 est ', occurences(plateau,1))
        print(' Le nombre de pions de joueur 2 est ', occurences(plateau,2))
    else:
        print(' Match nul ')
        
    print(' Le plateau final est ')
    afficherMatrice(plateau,tailleCellule)
        
#programme pour tester
nbLignes=8
nbColonnes=8
valeurParDefaut=0
tailleCellule=4
plateau=newMatrice(nbLignes, nbColonnes, valeurParDefaut)
setVal(plateau,3,3,1)
setVal(plateau,3,4,2)
setVal(plateau,4,3,2)
setVal(plateau,4,4,1)
partie(plateau, tailleCellule)

                        

                          


```
